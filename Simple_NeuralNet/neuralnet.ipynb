```python
# Create first network with Keras
from keras.models import Sequential
from keras.layers import Dense
import  numpy as  np
```

```python
# fix random seed for reproducibility
seed = 187
np.random.seed(seed)
import sklearn
from sklearn.preprocessing import normalize

# split into input (X) and output (Y) variables
means=np.mean(X,0)
stds=np.std(X,0)
X=(X-means)/stds
m=X.shape[0]
n=X.shape[1]


trainX=X[ 0: int(0.9*m)  ,  :]
trainY=Y[ 0: int(0.9*m)  , :]


testX=X[ int(0.9*m)  :m, :]
testY=Y[  int(0.9*m)  :m, :]

```

```python
# create model
x1=int(n/10)
model = Sequential()
model.add(Dense(x1, input_dim=n, init='normal', activation='relu'))
model.add(Dense(x1, init='normal', activation='relu'))
model.add(Dense(x1, init='normal', activation='relu'))
model.add(Dense(x1, init='normal', activation='relu'))

model.add(Dense(n, init='normal'))
# Compile model
```

```python

model.compile(loss='mean_squared_error', optimizer='adam')
# Fit the 
```

```python
model.fit(trainX, trainY, nb_epoch=1000, batch_size=100,  verbose=2)
# calculate predictions
```

```python
def func2(nv):
    inp=testX[nv:nv+1,:]
    prd = model.predict(inp)
    #print predictions.shape
    inp=inp.reshape((64,64,3))
    prd=prd.reshape((64,64,3))
    inp=inp*255
    prd=prd*255


    import scipy.misc
    scipy.misc.imsave('res_'+str(nv)+'.png', inp)
    scipy.misc.imsave('res_'+str(nv+1)+'.png', prd)

for i in range(0,100,2):
    func2(i)
func2(1)
```
